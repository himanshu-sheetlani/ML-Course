### 1. Basic Function Definition

```python
def funct1(fname):
  print(fname)
funct1("fchchg")
```

### 2. Function with One Argument

```python
def funct1(fname, lname ):
  print(fname + "  " +lname)
funct1("fchchg", "vsdvhsg")
```

### 3. Function with Multiple Arguments

```python
def funct1(*kid):
  print(kid[2])
funct1("a", "b", "c")
funct1("a", "b", "r", "d")
funct1("a", "b", "t", "d", "e")
```

### 4. Arbitrary Arguments (`*args`)

```python
def funct1( k,j,i):
  print(" the values of i = {}, j={}, k={}" .format(i,j,k))

funct1(i=2, j=6, k=8)
```

### 5. Keyword Arguments

```python
def funct1( ** var):
   print(var["i"])
funct1(i=2, j=6, k=8)
```

### 6. Arbitrary Keyword Arguments (`**kwargs`)

```python
def funct1(food):
  for i in food:
    print(i)

fruits=["apple", "banana", "cherry"]
funct1(fruits)
```

### 11. Local vs. Global Variables (without `global`)

```python
x=10
def funct1():
  x=globals()['x']
  x=10
  x= x*10
  globals()['x']=x
  print(x)

funct1()
print(x)
```

### 12. Modifying Global Variables using `globals()`

```python
x=10
def funct1():
  global x
  x= x*100
  print(x)

funct1()
print(x)
```
