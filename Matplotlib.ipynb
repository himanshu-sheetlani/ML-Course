### 3D Scatter Plot for Machine Learning Data

This plot visualizes synthetic 3D data points representing two classes. In machine learning, such plots are useful for understanding how different features separate data points into distinct categories. Here, `x`, `y`, and `z` could be three features, and the color represents the class label. This helps in visual assessment of class separability.

```python
#3D Scatter Plot for Machine Learning Data
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate synthetic data for two classes in 3D
np.random.seed(42)

# Class 0 data
x0 = np.random.rand(50) * 2
y0 = np.random.rand(50) * 2
z0 = np.random.rand(50) * 2

# Class 1 data (shifted)
x1 = np.random.rand(50) * 2 + 3
y1 = np.random.rand(50) * 2 + 3
z1 = np.random.rand(50) * 2 + 3

# Create a figure and a 3D axes object
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot Class 0
ax.scatter(x0, y0, z0, c='red', marker='o', label='Class 0')

# Plot Class 1
ax.scatter(x1, y1, z1, c='blue', marker='^', label='Class 1')

# Add labels and title
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('3D Scatter Plot of Synthetic Data for ML Classes')
ax.legend()

plt.show()
```

### Four Subplots in a 2x2 Grid

This example extends the subplot concept to a 2x2 grid, allowing for the visualization of four different relationships or aspects of data simultaneously within one figure.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
y_exp = np.exp(-x / 5) * np.sin(x * 5) # Damped sine wave

# Create a figure and a set of subplots arranged in a 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10)) # 2 rows, 2 columns

# Plot 1: Sine wave
axes[0, 0].plot(x, y_sin, color='red')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].grid(True)

# Plot 2: Cosine wave
axes[0, 1].plot(x, y_cos, color='blue', linestyle='--')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X-axis')
axes[0, 1].set_ylabel('Y-axis')
axes[0, 1].grid(True)

# Plot 3: Tangent wave (with limits for better visualization)
axes[1, 0].plot(x, y_tan, color='green', linestyle=':')
axes[1, 0].set_title('Tangent Wave')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')
axes[1, 0].set_ylim(-5, 5) # Limit y-axis for clarity
axes[1, 0].grid(True)

# Plot 4: Damped Sine Wave
axes[1, 1].plot(x, y_exp, color='purple', linestyle='-.')
axes[1, 1].set_title('Damped Sine Wave')
axes[1, 1].set_xlabel('X-axis')
axes[1, 1].set_ylabel('Y-axis')
axes[1, 1].grid(True)

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()
plt.show()
```

### Saving a Matplotlib Figure

To save a figure, you can use the `plt.savefig()` function after creating your plot. You can specify the filename and format (e.g., `.png`, `.jpg`, `.pdf`).

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a simple plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Save the figure to a file
# You can change the filename and format as needed
plt.savefig('sine_wave_plot.png')

plt.show()
print('Figure saved as sine_wave_plot.png')
```
