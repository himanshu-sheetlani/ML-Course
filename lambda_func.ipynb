```python
l=[1,2,3,4,5,6,7,8,9]
result=filter(lambda x: x%2==0, l) #filter returns object
print(list(result))
```

```python
l=[1,2,3,4,5,6,7,8,9]
cubed= map(lambda x: pow(x,3), l)
print(list(cubed))
```

```python
def add(n):
  return n+n
l=[1,2,3,4,5,6,7,8,9]
result=map(add,l)
print(list(result))
```

```python
l=['12345', 'abc','bat','cat','mat']
result=list(map(list, l))
print(result)
```

```python
cube=lambda x:x*x*x
print(cube(5))
```

```python
tables = [lambda x=x:x*10 for x in range(1,11)]
for table in tables :
  print(table())
```

```python
l=[1,2,3,4,5,6]
list_of_lambda=[lambda i=i: i*i for i in l]
# print((list_of_lambda))
for i in list_of_lambda:
  print(i())
```

```python
Max=lambda a,b : a if (a>b) else b
print(Max(4,2))
```

```python
l=[[2,3,4], [1,24,16,64], [3,6,9,12]]
s_list=lambda x:[sorted(i) for i in x]
print(s_list(l))
```

```python
l=[[2,3,4], [1,24,16,64], [3,6,9,12]]
s_list=lambda x:[sorted(i) for i in x]
print(s_list(l))
second_largest = lambda y: [y[-2] for y in s_list(l)]
print(second_largest(l))
```

```python
animal=["dog", "cat", "parrot", "rabbit"]
u_animal=list(map(lambda x: str.upper(x), animal))
print(u_animal)
```

```python
from functools import reduce
li = [5, 8, 10, 20, 30]
sum = reduce((lambda x, y: x + y), li)
print(sum)
```

```python
l=[1,3,5,6,2]
print("the max element of the list", reduce(lambda a,b: a if a>b else b, l))
```
