```python
class Car:
  def __init__(self,name,model,year):
    self.name=name
    self.model=model
    self.year=year
  def display_info(self):
    print("car: ", self.name,self.model,self.year)

car1=Car("Toyota","Camry",2022)
car2=Car("Honda","Civic",2021)
car1.display_info()
car2.display_info()
```

```python
class Student:
  def __init__(self, name, rollno):
    self.name=name
    self.rollno=rollno
  def print_info(self):
    print("name ", self.name, "rollno ", self.rollno)
s1=[Student("Himanshu", 48),
    Student("Harshita", 47),
    Student("XYZ", 50)
    ]
for i in s1:
  i.print_info()
```

```python
class Emp:
  def __init__(self, name, salary):
    self.name=name
    self.salary=salary
  def print_info(self):
    print("name ", self.name, "salary ", self.salary)
es1={
    101:Emp("Himanshu", 1200000),
    102:Emp("Harshita", 1200000),
    103:Emp("XYZ", 50)
}
for i in es1:
  es1[i].print_info()
```
